# 📝 Análise das Atas do Copom: Sentimentos Conservador e Flexível

Vamos explorar juntos as atas do Comitê de Política Monetária (Copom) para identificar e entender as nuances entre uma postura **Conservadora (Hawkish)** 🦅 e uma postura **Flexível (Dovish)** 🕊️. Essas duas classificações são fundamentais para compreender o direcionamento da política monetária do Banco Central e suas implicações na economia.

## 🔍 Por que isso é importante?

A análise do sentimento expressado nas atas do Copom nos ajuda a decifrar as intenções futuras do Banco Central em relação à taxa de juros e à política monetária. Isso é crucial para investidores, economistas e todos aqueles interessados em prever os próximos movimentos da economia.

### **Postura Conservadora (Hawkish) 🦅**
Um tom **conservador (hawkish)** sugere que o Banco Central está mais preocupado com a inflação e pode adotar medidas mais duras, como manter ou até aumentar a taxa de juros para conter a alta dos preços. Isso geralmente ocorre em cenários onde há sinais de inflação elevada e crescimento econômico acelerado, demandando uma postura mais rígida.

### **Postura Flexível (Dovish) 🕊️**
Por outro lado, um tom **flexível (dovish)** indica uma maior preocupação com o crescimento econômico e o emprego, levando o Banco Central a adotar uma postura mais flexível, com a possibilidade de cortes nas taxas de juros. Esse tipo de postura é mais comum quando a inflação está controlada e há sinais de desaceleração econômica.

## 🎯 Objetivo da Análise

Nosso objetivo é identificar os sinais que indicam se o Copom está adotando uma postura mais **conservadora (hawkish)** ou **flexível (dovish)** e entender como essas inclinações afetam a economia. Usando técnicas de análise de sentimento e interpretação, vamos mergulhar nas atas e extrair insights valiosos.

## 🌟 Importância para Diversas Áreas

A análise dos sentimentos nas atas do Copom vai muito além do interesse acadêmico. Ela é uma ferramenta essencial para **investidores**, que precisam antecipar movimentos nos mercados financeiros e ajustar suas estratégias de investimento; para **acadêmicos**, que estudam o impacto das políticas monetárias e contribuem para o avanço do conhecimento econômico; para **analistas econômicos**, que interpretam esses sinais para oferecer previsões precisas e recomendações sólidas; e para **gestores empresariais**, que dependem dessas análises para tomar decisões estratégicas, como alocação de recursos, planejamento financeiro e gestão de riscos.

```python
# Bibliotecas ----
from langchain_community.document_loaders import WebBaseLoader
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.prompts import PromptTemplate
from langchain.schema import Document, StrOutputParser
from langchain.schema.prompt_template import format_document
import os
import requests
import json

# Autenticação na API do Google
os.environ["GOOGLE_API_KEY"] = "AIzaSyBqg-zbG4b4tM_6u0X9G9mIZ_77NdCgMu8"  # Coloque aqui sua chave

# Definição do modelo LLM ----
llm = ChatGoogleGenerativeAI(model="gemini-pro", temperature=0)

# Engenharia de prompt ----
doc_prompt = PromptTemplate.from_template("{page_content}")

llm_prompt = PromptTemplate.from_template("""
Você é um economista chefe experiente em análise de política monetária, tendo
trabalhado em fundos de investimentos e grandes bancos, além de ser PhD em
economia.

Sua tarefa é ler a ata a seguir da Reunião do Comitê de Política Monetária (Copom)
do Banco Central do Brasil e classificar o documento em hawkish ou dovish e explicar os principais motivos desta decisão em 2 a 7 bullets. Dar uma
nota no final de intensidade de Hawkish ou Dovish (Baixo, Moderado ou Alto):

"{ata}"

Classificação:""")

# Função para coletar atas do Copom ----
def coletar_atas(quantidade=3):
    # Obter o número da última ata disponível
    ultima_ata = requests.get(
        "https://www.bcb.gov.br/api/servico/sitebcb/copom/atas?quantidade=1"
    ).json()["conteudo"][0]["nroReuniao"]

    url_ata = "https://www.bcb.gov.br/api/servico/sitebcb/copom/atas_detalhes?nro_reuniao="
    documentos = []

    # Coletar as atas de acordo com a quantidade especificada
    for num in range(ultima_ata - quantidade + 1, ultima_ata + 1):
        url_completa = url_ata + str(num)
        ata = WebBaseLoader(url_completa).load()
        conteudo = json.loads(ata[0].page_content)["conteudo"][0]
        documentos.append(
            Document(
                page_content=conteudo["textoAta"],
                metadata={"data": conteudo["dataReferencia"]}
            )
        )

    return documentos

# Função para classificar atas ----
def classificar_atas(documentos):
    resultados = []
    classificador = (
        {"ata": lambda doc: format_document(doc, doc_prompt)}
        | llm_prompt
        | llm
        | StrOutputParser()
    )
    for doc in documentos:
        resultado = classificador.invoke(doc)
        resultados.append({
            "data": doc.metadata["data"],
            "classificação": resultado
        })

    return resultados

# Coleta de atas e classificação ----
quantidade_atas = 3  # Exemplo: analisar as 3 últimas atas
documentos = coletar_atas(quantidade=quantidade_atas)
resultados = classificar_atas(documentos)

# Exibir resultados
for resultado in resultados:
    print(f"Ata de {resultado['data']}:")
    print(resultado['classificação'])
    print("\n" + "-"*50 + "\n")
```
